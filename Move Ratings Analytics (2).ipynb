```python
#Styling Dashboard
```

```python
sns.set_style("dark", {"axes.facecolor": "black"})
f, axes = plt.subplots(2,2, figsize=(15,15))
#Plot 0,0
k1 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating', shade=True, shade_lowest=False, cmap = 'inferno', levels=9, ax = axes[0,0])
k1b = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating', cmap = 'cool', levels=9, ax = axes[0,0])

#Plot 0,1
k2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating', shade=True, shade_lowest=False, cmap = 'inferno', ax = axes[0,1])
k2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating', cmap = 'cool', ax = axes[0,1])

#Plot 1,0
z = sns.violinplot(data = movies, x ='Year', y = 'BudgetMillions', palette='YlOrRd', ax = axes[1,0])

#Plot 1,1
k4 = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', shade = True, cmap = 'Blues_r', ax = axes[1,1])
k4b = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', cmap = 'gist_gray_r', ax = axes[1,1])


k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
list1 = list()
mylabels = list()
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillions)
    mylabels.append(gen)
    
sns.set_style('whitegrid')
fig, ax = plt.subplots() #by adding it in subplot, it is easier to change the size
fig.set_size_inches(11.7,8.27) #size of A4 paper
h = plt.hist(list1, bins = 30, stacked = True, rwidth=1, label = mylabels, color = ["Red","Green","Yellow","Orange","Blue","Black","Grey"])
plt.title("Movie Budget Distribution", fontsize=35, color='DarkBlue',fontname='Console')
plt.ylabel("Number of Movies", fontsize=25, color='Red')
plt.xlabel("Budget", fontsize=25, color='Green')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(frameon=True, fancybox=True, shadow=True, framealpha=1, prop={'size':10})
plt.show()
```
